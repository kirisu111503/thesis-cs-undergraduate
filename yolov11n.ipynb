```python
import os
import zipfile
from ultralytics import YOLO

# 1. Download the file from Google Drive
# The ID is extracted from your link: 11MgeU5PHxRrbTR9Q-fHUt82a55a9-n5X
!gdown 11MgeU5PHxRrbTR9Q-fHUt82a55a9-n5X -O dataset.zip

# 2. Extract the dataset
print("Extracting dataset... please wait.")
with zipfile.ZipFile('dataset.zip', 'r') as zip_ref:
    zip_ref.extractall('/content/')
print("Extraction complete.")

# 3. Create the dataset.yaml file automatically
yaml_content = f"""
path: /content/dataset
train: train/images
val: val/images
test: test/images

names:
  0: Phone
  1: Calculator
  2: Book
"""

with open('/content/dataset.yaml', 'w') as f:
    f.write(yaml_content)

print("dataset.yaml created successfully.")
```

```python
# Load the YOLOv11n model (it will auto-download the weights)
model = YOLO('yolo11n.pt')

# Start training
# We use 'device=0' to ensure it uses the Colab T4 GPU
results = model.train(
    data='/content/dataset.yaml',
    epochs=200,        # High limit
    patience=30,       # If no improvement for 50 epochs, stop early
    imgsz=640,
    batch=16,
    device=0,
    name='yolov11_gadgets_final'
)
```

```python
import shutil
from google.colab import files

# 1. Define the results folder path
# Note: 'yolov11_gadgets' is the name we used in the model.train() command
results_path = '/content/runs/detect/yolov11_gadgets_final'
zip_name = 'yolov11_training_results.zip'

# 2. Create a zip file of the entire results directory
print(f"Zipping results from {results_path}...")
shutil.make_archive('yolov11_training_results', 'zip', results_path)

# 3. Download the zip file to your computer
print("Downloading results...")
files.download('yolov11_training_results.zip')
```
